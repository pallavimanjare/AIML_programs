```python
# Decision Control Structures | Conditional Statement
#
# 1. if statement
# 2. if ..else statement
# 3. elif ladder
```

```python
# if statement
#
# In this expression we can supply one condition. If the condition is met, it will execute the logic coded
#
# Syntax:
#
# if condition:
#     statement
```

```python
#Write a program to accept number from user and check if the given number is greater than 10

num1 = int(input("Enter number: "))

if num1 > 10:
  print(f"Given number {num1} is greater than 10")
```

```python
#if ... else statement
#
# In this expression, we can supply BINARY CONDITION (Two conditions only)
#
# Syntax:
#
# if condn1:
#      statement1
# else:
#      statement2
```

```python
#Write a program to accept number from user and check if the given number is greater than 10

num1 = int(input("Enter number: "))

if num1 > 10:
  print(f"Given number {num1} is greater than 10")
else:
  print(f"Given number {num1} is less than or equal to 10")
```

```python
# elif ladder
#
# In this expression you can supply multiple conditions
#
#

num1 = int(input("Enter number: "))

if num1 > 10:
  print(f"Given number {num1} is greater than 10")
elif num1 == 10:
  print(f"Given number {num1} is equal to 10")
else:
  print(f"Given number {num1} is less than or equal to 10")
```

```python
# Accept the number from the user and display message based on given conditions
# Number between 1 to 10 -----> Good
# Number between 11 to 20 ----> Better
# Number between 21 to 30 ----> Best
# Number above 30 ------------> Outstanding


#Defining a series or range

#11 <= num1 <= 20
```

```python
num1 = int(input("Enter number: "))

if num1 >= 1 and num1 <= 10:
  print("Good")
elif num1 >= 11 and num1 <= 20:
  print("Better")
elif num1 >= 21 and num1 <= 30:
  print("Best")
else:
  print("Outstanding")
```

```python
num1 = int(input("Enter number: "))
if num1 <= 10:
    print("Good")
elif 11 <= num1 <= 20:
    print("Better")
elif 21 <= num1 <= 30:
    print("Best")
elif num1 > 30:
    print("Better")
```

```python
#Question: Loan App Example

# Approve the loan of the customer if the sal is greater than 8000 and the credit score is greater than 700
# The applicant must not be a student
# if the applicant is a student reject the loan irrespective of the sal and credit condition

#User input

# Enter Salary:
# Enter Credit Score (0-800):
# Are you a student (Y/N):
```

```python
salary = int(input("Enter Salary: "))
credit_score = int(input("Enter Credit Score (0-800): "))
is_student = input("Are you a student (Y/N): ").strip().upper()
if is_student == "Y":
  print("Loan Rejected!")
elif salary > 8000 and credit_score > 700:
  print("Loan Approved!")
else:
  print("Loan Rejected!")
```
