```python
#Exception Handling
#
# The goal here is to make the program robust and error-resistant
#
# 1. Try--except block
# 2. The else block
# 3. Finally block
```

```python
10 / 0
```

```python
num1 = int(input("Enter first number: "))
num2 = int(input("Enter second number: "))

result = num1 + num2

print(f"Addition of {num1} and {num2} is {result}")
```

```python
try:
  10 / 0
except:
  print("Cannot divide by zero")
```

```python
try:
  10 / 0
except ZeroDivisionError as e:
  print(e)
```

```python
try:
  num1 = int(input("Enter first number: "))
  num2 = int(input("Enter second number: "))

  result = num1 + num2

  print(f"Addition of {num1} and {num2} is {result}")

except:
  print("Please enter numeric value")
```

```python
#Using else
# Here else means the statements/block inside else will execute only when there is no exception encountered

try:
  num1 = int(input("Enter first number: "))
  num2 = int(input("Enter second number: "))
except:
  print("Enter numbers only")
else:
  result = num1 + num2
  print(f"Addition of {num1} and {num2} is {result}")

```

```python
#finally block ---- Execute the code irrespective of errors

try:
  num1 = int(input("Enter first number: "))
  num2 = int(input("Enter second number: "))
except:
  print("Enter numbers only")
else:
  result = num1 + num2
  print(f"Addition of {num1} and {num2} is {result}")

finally:
  print("Thanks for using my program")
```

```python
try:
  num1 = int(input("Enter first number: "))
  num2 = int(input("Enter second number: "))
except:
  print("Enter numbers only")
else:
  try:
    result = num1 / num2
  except:
    print("Divide by zero not allowed")
  else:
    print(f"Division of {num1} and {num2} is {result}")

finally:
  print("Thanks for using my program")
```
