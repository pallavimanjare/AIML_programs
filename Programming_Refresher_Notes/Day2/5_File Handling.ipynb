```python
#File Handling
# 1. Read a file
# 2. Write content in a file
```

```python
# Built in function called open()

# Syntax:

# open(fileLocation, mode)
#
# modes:
#
# r ------ read
# w ------ write
# a ------ append
# r+ ----- reading and writing
```

```python
#Write a file

with open("abc.txt",'w') as f:
  f.write("Welcome to file handling concept")
```

```python
#Read a file

with open("abc.txt", 'r') as f:
  for line in f:
    print(line)
```

```python
#Append data in existing file

with open("abc.txt",'a') as f:
  f.write("\nThis is the third data")
```

```python
list1 = ["Welcome to Python\n","This is file handling\n"]
with open("abc.txt",'w') as f:
  f.writelines(list1)
```
