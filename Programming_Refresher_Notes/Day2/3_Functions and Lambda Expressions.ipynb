```python
def factorial(number = 1):
  result = 1
  for num in range(1, number + 1):
    result = result * num

  return result
```

```python
factorial(5)
```

```python
#Important Preprocessing and Transformation (Data Analysis and Data Transformation)
# 1. map
# 2. Filter
```

```python
#map function --- This allows you to map a logic to an iterable object (collection object)

#Syntax: map(functionLogic,iterableObj)

list1 = [1,2,3,4,5]

#Goal --- Square each element in the given list

#Solution1:

resultList=[]
for num in list1:
  resultList.append(num ** 2)

resultList
```

```python
#Solution2: Storing logic in function and reusing logic using map

def square(num):
  return num ** 2
```

```python
list(map(square,list1))
```

```python
# 2. Filter ---- Allows to filter elemnents that satisfy logic (True boolean values are returned)

#Always create function that returns boolean output
def isEven(num):
  return num % 2 == 0

list1=[1,2,3,4,5]

list(filter(isEven,list1))
```

```python
list(map(isEven,list1))
```

```python
def greaterThanFour(num):
  if num > 4:
    return True

list1=[1,2,3,4,5]

list(filter(greaterThanFour,list1))
```
