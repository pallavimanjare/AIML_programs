```python
#1. Creating Strings

#Single Quote

string1 = 'Hello world'
print(string1)

#Double Quote

string2 = "Hello world"
print(string2)

#Triple Quote

string3 = """

      Welcome to Python Class
      -----------------------

      We are learning
        a. String Handling
        b. For loop
        c. Functions
        d. File Handling Basics
        e. Exception Handling Basics

"""

print(string3)
```

```python
# 2. Accessing characters / substrings in a string variable

testData = "Welcome to Python Class"

#Accessing single character

print(testData[11])
```

```python
#To extract collection of characters ---- Using the Slicing Method

print(testData[11:17])
```

```python
#Basic String Operations

#Concatenation ------- Merging of two or more strings (+)

var1 = "Hello"
var2 = "World"

print(var1+var2)
print(var1+" "+var2)
```

```python
#Repetition (*)

print(var1 * 3)
temp = var1+" "
print(temp * 3)
```

```python
#Search Operation --- Check if a word exists in a given string object
# in operator

testData = "Welcome to Python Class"
searchKeyword = input("Enter value to search: ")

#Check if the word Python exists in my testData dataset

searchKeyword.lower() in testData.lower()
```

```python
#Basic String Methods

testData = "Welcome to Python Class"

# 1. Change Case methods -----> Normalization

print(testData.lower())
print(testData.upper())
```

```python
testData = "Welcome to Python Class"

#Check if the word Python exists in my testData dataset

"Python".lower() in testData.lower()
```

```python
txt = "hello, and welcome to my world."

x = txt.capitalize()

print (x)
```

```python
#Find and Replace

#find() ---- returns the startIndex location of the given word

data = "Welcome to the world of Python. You are learning string handling in Python with Prashant Nair"

position = data.find("Python")
print(position)
```

```python
#replace() ----- Replace a given source string with given destination string

updatedString = data.replace("Python","Java")
print(updatedString)
```

```python
#Stripping Whitespace

test = "  Hello  "

print(test.strip())
print(test.lstrip())
print(test.rstrip())
```
