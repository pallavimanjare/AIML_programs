```python
#Create a function that can accept salary and return the bonus of the salary
# salary less than or equal to 1500 ------------> 10% bonus
# salary between 1501 and 6000 -----------------> 5% bonus
# salary between 6001 and 9900 ----------------> 2.5%
# salary is greater than 9900 -----------------> 0

def bonus(salary):
  if salary <= 1500:
    return round(salary * 0.10,3)
  elif salary > 1500 and salary <= 6000:
    return round(salary * 0.05,3)
  elif salary > 6000 and salary <= 9900:
    return round(salary * 0.025,3)
  else:
    return 0
```

```python
bonus(9878)
```

```python
#Pandas allow to apply function as a broadcast in a column

#Create a new column called bonus and calc bonus of each employee based on his/her monthly salary

#apply()

empWithHeadersDF['bonus'] = empWithHeadersDF['esal'].apply(bonus)

empWithHeadersDF
```
