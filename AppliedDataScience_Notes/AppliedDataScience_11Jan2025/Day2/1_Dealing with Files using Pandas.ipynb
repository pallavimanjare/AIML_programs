```python
#How to create a datafreame using a collection object?

list1 = [
    [1,'Prashant',1000],
    [2,'Arun',2000]
]
```

```python
# DataFrame contains two types of index
# Row Index
# Column Index

#Usually from analysis perspective, it is recommended to have column names which will simplify your querying process


empFromList.columns = ['empid','empname','empsalary']
empFromList
```
