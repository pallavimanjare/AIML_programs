```python
#Concatenation

# pd.concat -----> default is rowwise contactentation

df1 = pd.DataFrame({
    'EmployeeID': [1, 2, 3, 4],
    'Name': ['John', 'Jane', 'Tom', 'Lucy'],
    'Department': ['HR', 'Finance', 'IT', 'Marketing']
})

df3 = pd.DataFrame({
    'EmployeeID': [5,6],
    'Name': ['John', 'Lucy'],
    'Department': ['Finance', 'IT']
})
```

```python
concatDF = pd.concat([df1,df3])
concatDF
```

```python
concatDF.reset_index(drop=True)
```

```python
#Method 2
concatDF = pd.concat([df1,df3], ignore_index=True)
concatDF
```

```python
import pandas as pd

# Creating a sample DataFrame
data = {
    'Product': ['A', 'B', 'A', 'B', 'C', 'A', 'C', 'B'],
    'Units_Sold': [150, 200, 300, 400, 250, 100, 300, 500],
    'Revenue': [4500, 6000, 9000, 12000, 7500, 3000, 9000, 15000],
    'Cost': [3000, 4000, 5000, 7000, 3500, 2000, 6000, 10000]
}
df = pd.DataFrame(data)
print("Sample DataFrame:")
print(df)
```

```python
#To get the unique values of the given column

df['Product'].unique()
```

```python
#Get the total records for each unique value

df['Product'].value_counts()
```

```python
#Group by Single Column
df.groupby('Product').sum()
```

```python
#Group by Multiple Columns

df.groupby(['Product','Cost']).sum()
```
