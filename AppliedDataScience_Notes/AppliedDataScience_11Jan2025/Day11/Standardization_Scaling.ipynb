```python
# Feature Preparation
# 1. Standardization --- Its a data preprocessing technique used to transform the data such that the mean of the variable is ZERO and standard
#                        deviation is 1  (Z-score Normalization)
```

```python
from sklearn.preprocessing import StandardScaler

#Sample Dataset
data = [[1,200],[2,300],[3,400],[4,500],[5,600]]

#initailize the scaler object

sc = StandardScaler()

#
sc.fit_transform(data)
```

```python
import numpy as np
data2 = np.array([[100],[200],[300],[400],[500]])
sc2 = StandardScaler()
transformedData = sc2.fit_transform(data2)
```

```python
import seaborn as sns
sns.displot(data=data2, kind="kde")
```

```python
import seaborn as sns
sns.displot(data=transformedData, kind="kde")
```

```python
#Other Scaling Techniques

# 1. MinMaxScaling
# 2. Robust Scaling
```

```python
#MinMax Scaling ---> Data is scaled based on the range supplied by the user (default range is 0 to 1)

from sklearn.preprocessing import MinMaxScaler

data2 = np.array([[100],[200],[300],[400],[500]])
mm = MinMaxScaler()
transformedData2 = mm.fit_transform(data2)
```

```python
from sklearn.preprocessing import MinMaxScaler

data2 = np.array([[100],[200],[300],[400],[500]])
mm = MinMaxScaler(feature_range=(1,2))
transformedData2 = mm.fit_transform(data2)
```

```python
transformedData2
```

```python
import seaborn as sns
sns.displot(data=transformedData2, kind="kde")
```

```python
# 2. Robust Scaling --- Another Scaling Technqiue that scales the data that are robust to outliers
#

from sklearn.preprocessing import RobustScaler

data2 = np.array([[100],[200],[300],[400],[500000000]])
mm = RobustScaler()
transformedData3 = mm.fit_transform(data2)
```

```python
transformedData3
```

```python
sns.displot(data2,kind='kde')
```

```python
sns.displot(transformedData3,kind='kde')
```
