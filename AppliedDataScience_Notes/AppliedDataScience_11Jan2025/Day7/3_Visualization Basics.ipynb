```python
import seaborn as sns
```

```python
tips = sns.load_dataset("tips")
```

```python
tips.info()
```

```python
tips.head()
```

```python
#Two types of data column in broad spectrum
# 1. Numerical Data
#      a. Continuous ND ()
#      b. Discrete ND
# 2. Non-numerical Data
#      a. Categorical Data
#      b. Pure String data
```

```python
#Continuous ND ------------
#-> Use Distribution Plots
#-> Use Box Plots (Outlier Detection)
#-> Use joint plots (Correlation Analysis)

sns.distplot(tips['total_bill'])
```

```python
tips.describe()
```

```python
#Plot ---- Box plot

sns.boxplot(x=tips['total_bill'])
```

image.png

```python
#Joint PLot

sns.jointplot(x=tips['tip'], y=tips['total_bill'])
```

```python
sns.jointplot(x=tips['tip'], y=tips['total_bill'], kind="reg")
```

```python
#Pair Plot

sns.pairplot(tips)
```
