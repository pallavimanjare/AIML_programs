```python
# Ordinal Data -- Categorical data with ranking
# Ordinal data represent categories with a meaningful order
# We need a way to represent the data in numerical form ---------------------> ORDINAL ENCODING
# Any kind of Encoding always starts from 0
```

```python
import pandas as pd
import numpy as np
```

```python
# Sample Data
data = {'Satisfaction': ["Neutral", "Satisfied", "Very Dissatisfied", "Very Satisfied", "Dissatisfied"]}
df = pd.DataFrame(data)
```

```python
df
```

```python
#Domain Expert
# Ordering from low to high
# 1. Very Dissatisfied
# 2. Dissatisfied
# 3. Neutral
# 4. Satisfied
# 5. Very Satisfied
```

```python
#Method1: Pandas Method --- Manual Mapping using map

mapping = {"Very Dissatisfied": 0,
           "Dissatisfied":1,
           "Neutral":2,
           "Satisfied":3,
           "Very Satisfied": 4}

df['Satisfaction_Encoded'] = df["Satisfaction"].map(mapping)
```

```python
df
```

```python
# Method 2: Using sci-kit learn
# You can use preprocessing package of sci-kit learn to encode ordinal data

from sklearn.preprocessing import OrdinalEncoder

#The below list must be in ascending order of rank
satisfactionOrder = ["Very Dissatisfied", "Dissatisfied", "Neutral", "Satisfied","Very Satisfied"]

encoder = OrdinalEncoder(categories=[satisfactionOrder])

df['Satisfaction_Encoded2'] = encoder.fit_transform(df[['Satisfaction']])

df
```
