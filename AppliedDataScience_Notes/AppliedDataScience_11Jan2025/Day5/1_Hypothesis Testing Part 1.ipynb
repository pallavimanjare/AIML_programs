```python
# Correlation Test

#Perform Hypothesis Testing
#
#
# 1. Create a Viable Question Goal (Binary Outcome)
#
#  Check if there exists any linear relationship between RDSpend and Profit
#
#

# 2. Convert the Question into Hypothesis (H0 and H1)
#
#   Null Hypothesis (H0) --------- RDSpend and Profit have NO linear relationship
#   Alternate Hypothesis (H1) ---- RDSpend and Profit have Linear Relationship
#
#

# 3. Select the Statistical test/formula/tool/method to validate the Hypothesis (Who wins?)
#
#   Correlation Test ---- Pearson's Correlation test formula

# 4. Select/Determine/Decide the SL of the project
#
#    SL = 0.05
#

# 5. Calc the p-value from the test and compare the same with the SL to determine who wins?
#
#

SL = 0.05


from scipy.stats import pearsonr

_,pvalue = pearsonr(data['R&D Spend'], data['Profit'])

if pvalue <= SL:
  print("Alternate Hypothesis (H1) ---- RDSpend and Profit have Linear Relationship")
else:
  print("Null Hypothesis (H0) --------- RDSpend and Profit have NO linear relationship")
```
