```python
import pandas as pd
import numpy as np
import seaborn as sns

#load csv file
melbDat = pd.read_csv('melb_data.csv')
```

```python
#2. Check for missing data and repair the same
melbDat.info()
```

```python
#It seems BuildingArea   YearBuilt   CouncilArea     have null values
#Building Area : Numerical :Discrete:Range
#Years Build : Numerical :Continues: Mean
#CouncilArea; Numerical : Mode First
melbDat['YearBuilt'].mean()
melbDat['YearBuilt'].fillna(melbDat['YearBuilt'].mean() , inplace= True)

melbDat['BuildingArea'].median()
melbDat['BuildingArea'].fillna(melbDat['BuildingArea'].median() , inplace= True)

melbDat['CouncilArea'].mode()
melbDat['CouncilArea'].fillna(melbDat['CouncilArea'].mode()[0] , inplace= True)
melbDat
```
